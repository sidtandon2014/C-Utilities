## Bivariate Analysis

```python
def cramers_V(var1,var2) :
    from scipy.stats import chi2_contingency
    '''crosstab =np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) # Cross table building
    stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
    obs = np.sum(crosstab) # Number of observations
    mini = min(crosstab.shape)-1 # Take the minimum value between the columns and the rows of the cross table
    return (stat/(obs*mini))
    '''
    
    conf_matrix = pd.crosstab(var1, var2)
    chi2 = chi2_contingency(conf_matrix)[0]
    n = sum(conf_matrix.sum())
    phi2 = chi2/n
    r,k = conf_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    result=np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))
    return result

#cramers_V(dataset["EmployeeLeft"],dataset["National Origin (code)"])
```

```python
association = pd.DataFrame(columns=['VarName','Target',"Assoc"])
targetCol = "EmployeeLeft"
for col in dataset.columns:
    if col != targetCol:
        pd.crosstab(dataset[col], dataset[targetCol]).plot.bar()
        assoc = cramers_V(dataset[targetCol], dataset[col])
        association = association.append({"VarName":col,"Target":targetCol,"Assoc":assoc},ignore_index=True)
#association
```

## Observations

- **Nation Origin (code)**: Employee attrition is high in certain groups of people 
- **Salary level relative to Peers:**: As the difference between salary increases data is showing high chances of employee leaving
- **Survey, Relative, Attitude toward Peers**: As the difference between peers grow, high chances of employee leaving
- **Survey, Relative, Attitude toward Resources**: Some of the leaving employee are not satisfied with the resources provided
- **Survey, Relative, Attitude toward Work type**: People are not satisfied with the type of work 
- **Survey, Relative, Attitude toward Work load**: Certain employees feel the heat of workload
