```python
import pandas as pd
import pandas_profiling
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier 
```

```python
dataset = pd.read_csv("../Data/siddharth_Microsoft_issueDataset.csv")
```

```python
independentVars = [col for col in dataset.columns if col != target]
```

```python
from sklearn.model_selection import train_test_split
target = "EmployeeLeft"
independentVars = [col for col in dataset.columns if col != target]
Y = dataset[target]
X = dataset[independentVars]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
estimator = DecisionTreeClassifier(max_depth = 5)
estimator.fit(X_train,y_train)

test_predict = estimator.predict(X_test)
```

```python
test_predict
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, test_predict)
```

```python
from sklearn import tree
text_representation = tree.export_text(estimator)
print(text_representation)
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(estimator, 
                   feature_names=independentVars,  
                   class_names=target,
                   filled=True)
```
